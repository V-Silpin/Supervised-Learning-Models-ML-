```python
# Z score Method  (defines how many standard deviations away the data is from the mean)
# Bell Curve
df['zscore'] = (df.sepal_width - df.sepal_width.mean())/df.sepal_width.std()
df.head()
```

```python
df.sepal_width.hist()
mb.pyplot.show()
```

```python
print("The outliers are: ")
df[(df.zscore < -2) | (df.zscore > 2)]
```

```python
df1 = df[(df.zscore > -2) & (df.zscore < 2)]
df1
```

```python
df.zscore.hist()
mb.pyplot.show()
```

```python
df1.zscore.hist()
mb.pyplot.show()
```

```python
sb.boxplot(df1)
```

```python
# Multi Class
df1.groupby('species').size()
```

```python
sb.heatmap(df1.corr(), annot = True)
```

```python
#Encode Object items
from sklearn.preprocessing import LabelEncoder
le  = LabelEncoder()
df1['species'] = le.fit_transform(df1['species'])
df1
```

```python
# Find columns which have strong correlation (±0.80)
#(+ corr, - corr)
sb.heatmap(df1.corr(), annot = True)
```

```python
feature_columns = ['sepal_length','petal_length','petal_width']
X = df1[feature_columns].values
Y = df1.species.values
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train,Y_train)
```

```python
knn.predict(X_test)
```

```python
knn.score(X_test,Y_test)*100
```

```python
from sklearn.model_selection import GridSearchCV
knn2 = KNeighborsClassifier()
param = {"n_neighbors" : np.arange(1,35)}
grid_knn2 = GridSearchCV(knn2, param_grid = param, scoring = 'accuracy', cv = 5)
```

```python
grid_knn2.fit(X_train,Y_train)
```

```python
grid_knn2.best_params_
```

```python
best_score = grid_knn2.best_score_
best_score
```

```python
grid_knn2.score(X_test,Y_test)
```

```python
#Grid Search train data is split into train & test
from pickle import dump
dump(grid_knn2, open("KNN_MODEL.sav",'wb'))
```
